## Tipos de requests HTTP (o consulta)

1. GET: Envía parámetros en la URL EJ: https://www.google.com/search?q=zapallo 
    * URL https://www.google.com/
    * Ruta: /search/
    * Parámetro: q
    * Valor Parámetro: "zapallo"
    
    
2. POST Similar a GET pero envía parámetros a travéz de un body (cuerpo) usualmente json

```
{
    "parametro": "valor"
}

```
    
3. PUT
4. DELETE

Para más informacion leer [aquí](https://www.tutorialspoint.com/http/http_requests.htm)

## Parámetros con python
Los siguientes ejemplos muestran como enviar parámetros en un request GET y en uno POST

### GET

````python
payload = {'q': 'hola'}
r = requests.get("http://www.google.com/search", params=payload)
````

````python
r.url
````

````python
r.text
````

### POST

````python
import json

url = 'https://api.github.com/some/endpoint'
payload = {'some': 'data'}
headers = {'content-type': 'application/json'}

r = requests.post(url, data=json.dumps(payload), headers=headers)
````

````python
r.text
````

## Que es una URL

URL significa Uniform Resource Locator. Es como un indice pero que apunta a una ubicación en internet específica. Hacia a donde aputa es configurable.

¿Que dirección específica aputa una URL? Aputa a una IP.
* EJ IP: 123.31.12.31 
* PUERTO STANDARD: 80
* PUERTO STANDARD SEGURO: 443

Ejemplo de IP+Puerto:
* 186.11.89.81:443

# Que es un servidor

Un servidor es un programa que admite y procesa requerimientos a travez de una red de computadores. En particular, un servidor HTTP admite y procesa requerimientos HTTP.

Tradicionalmente los servidores servian un archivo HTML y las distintas rutas de nuestra página mostraban distintos archivos HTML. Esto era muy costoso en terminos de desarrollo pues significaba que cada ruta necesitaba una página web distinta. 

Actualmente, servidores HTTP como flask generan dinámicamente el contenido de una página web usando un router. Estos routes manejan (handle) las distintas rutas y generan el conenido de la página web de forma dinámica, ahorrandole al desarrollador el trabajo de crear manualmente un archivo HTML para cada ruta.

### Magics de jupyter notebooks

Los notebook nos permiten usar comandos llamados magics para interactuar con programas externos a la aplicación de jupyer notebooks. En particular, para este ejemplo necesitaremos los siguientes dos

* Nota: Para cargar un archivo externo a jupyter puedo hacerlo con el magic `%load _filename_`
* Nota 2: Para guardar modificaciones podemos usar `%%writefile _filename_`

Además, usando el prefijo `!` podemos ejecutar comandos en una linea de comandos tradicional como bash.

````python
# %load app.py
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hola():
    return "Hola"

if __name__ == "__main__":
    app.run()
````

Una vez guardado el archivo app.py, podemos ejecutarlo desde la terminal con el comando 
```
python app.py
```

Una vez que esté corriendo, notaremos que por default, la aplicación se sirve en el localhost (nuestra propia máquina) en el puerto 5000. Podemos hacer el requerimiento desde python para obtener el resultado del servidor o bien abrir la url:puerto en un browser para obtener la respuesta.
